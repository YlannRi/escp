```python


# Load the dataset (Adjust the file path)
file_path = r"C:\Users\ylann\vscode\Life-Expectancy-data.xlsx"  # Adjust this to your file's location
data = pd.read_excel(file_path, header=0)

# Define column names as separate elements
column_names = ["Country", "Year", "Status", "Life expectancy", "Alcohol", "Hepatitis B", 
                "Measles", "BMI", "Polio", "Diphtheria", "HIV/AIDS", "thinness  1-19 years"]

# Assign column names to the DataFrame
data.columns = column_names

# Convert numeric columns to numeric data type
numeric_columns = ["Year", "Life expectancy", "Alcohol", "Hepatitis B", 
                   "Measles", "BMI", "Polio", "Diphtheria", "HIV/AIDS", "thinness  1-19 years"]

data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Print basic descriptive statistics
basic_stats = data.describe()
print("Basic Descriptive Statistics:")
print(basic_stats)

# Additional detailed analysis
# Maximum values
max_values = data.max(numeric_only=True)
print("\nMaximum Values:")
print(max_values)

# Minimum values
min_values = data.min(numeric_only=True)
print("\nMinimum Values:")
print(min_values)

# Average (mean) values
mean_values = data.mean(numeric_only=True)
print("\nAverage (Mean) Values:")
print(mean_values)

# Check for significant distributions among numeric features
# Plotting histograms for all numeric columns
data.hist(figsize=(16, 12), bins=20)
plt.tight_layout()
plt.show()

# Plotting density plots (KDE) for all numeric columns to understand the distribution better
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data[column], fill=True)
    plt.title(f'Density Plot for {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python


# Load the dataset (Adjust the file path)
file_path = r"C:\Users\ylann\vscode\Life-Expectancy-data.xlsx"
  # Adjust this to your file's location
data = pd.read_excel(file_path, header=0)

# Define column names as separate elements
column_names = ["Country", "Year", "Status", "Life expectancy", "Alcohol", "Hepatitis B", 
                "Measles", "BMI", "Polio", "Diphtheria", "HIV/AIDS", "thinness  1-19 years"]

# Assign column names to the DataFrame
data.columns = column_names

# Convert numeric columns to numeric data type
numeric_columns = ["Year", "Life expectancy", "Alcohol", "Hepatitis B", 
                   "Measles", "BMI", "Polio", "Diphtheria", "HIV/AIDS", "thinness  1-19 years"]

data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Print basic descriptive statistics
basic_stats = data.describe()
print("Basic Descriptive Statistics:")
print(basic_stats)

# Additional detailed analysis
# Maximum values
max_values = data.max(numeric_only=True)
print("\nMaximum Values:")
print(max_values)

# Minimum values
min_values = data.min(numeric_only=True)
print("\nMinimum Values:")
print(min_values)

# Average (mean) values
mean_values = data.mean(numeric_only=True)
print("\nAverage (Mean) Values:")
print(mean_values)

# Check for significant distributions among numeric features
# Plotting line plots for all numeric columns by year
for column in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Year', y=column, data=data)
    plt.title(f'Trend of {column} Over Years')
    plt.xlabel('Year')
    plt.ylabel(column)
    plt.show()
```
